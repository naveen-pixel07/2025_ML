```python
X = df[['Years of Experience', 'Education Level', 'Job Title', 'Age']]
y = df['Salary']

X = pd.get_dummies(X, columns=['Education Level', 'Job Title'], drop_first=True)
display(X.head())
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
import pandas as pd

model = LinearRegression()


imputer_X = SimpleImputer(missing_values=np.nan, strategy='mean')
X_imputed = imputer_X.fit_transform(X)

imputer_y = SimpleImputer(missing_values=np.nan, strategy='mean')
y_imputed = imputer_y.fit_transform(y.values.reshape(-1, 1)).ravel()


model.fit(X_imputed, y_imputed)


sample_input = pd.DataFrame({
    'Years of Experience': [5.0],
    'Education Level': ['Bachelor\'s Degree'],
    'Job Title': ['Software Engineer'],
    'Age': [28.0]
})


sample_input_encoded = pd.get_dummies(sample_input, columns=['Education Level', 'Job Title'], drop_first=True)


missing_cols = set(X.columns) - set(sample_input_encoded.columns)
missing_df = pd.DataFrame(0, index=sample_input_encoded.index, columns=list(missing_cols))
sample_input_encoded = pd.concat([sample_input_encoded, missing_df], axis=1)


sample_input_encoded = sample_input_encoded[X.columns]


imputer_X_sample = SimpleImputer(missing_values=np.nan, strategy='mean')
sample_input_imputed = imputer_X_sample.fit_transform(sample_input_encoded)



predicted_salary = model.predict(sample_input_imputed)


sample_true_salary = 70000.0


print("Sample Input:")
display(sample_input)
print("\nPredicted Salary:", predicted_salary[0])
print("Sample True Salary:", sample_true_salary)
print("Prediction Error:", abs(predicted_salary[0] - sample_true_salary))
display(X.head())
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

model = LinearRegression()

imputer_X = SimpleImputer(missing_values=np.nan, strategy='mean')
X_imputed = imputer_X.fit_transform(X)

imputer_y = SimpleImputer(missing_values=np.nan, strategy='mean')
y_imputed = imputer_y.fit_transform(y.values.reshape(-1, 1)).ravel()

scores = cross_val_score(model, X_imputed, y_imputed, cv=5, scoring='r2')
print("R-squared scores for each fold:", scores)
print("Mean R-squared across folds:", scores.mean())
```
